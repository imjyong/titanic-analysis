## **Groupby**

- 집단 별 요약 = 내가 원하는 (관심 있는) 집단의 요약한 값을 구할 때 사용, 그룹으로 묶을 기준 필요하며 그 기준에 따른 요약 통계치 필요
---
- 1. groupby 로 컬럼 묶어보기
  - 데이터프레임.groupby(내가 관심있는 컬럼)
- 2. 2개 이상 groupby 컬럼 진행
  - 데이터프레임.groupby(['컬럼1', '컬럼2'])
---
- 내가 원하는 데이터프레임만 추출하고 싶을 때
  - get_group(내가 groupby한 컬럼의 값)
---
- **df.groupby('내가 묶을 컬럼들')['내가 통계치를 보고 싶은 컬럼'].기초통계문법**
- groupby로 묶은 후 내가 원하는 요약 통계치 보고 싶을 때
- case 1
  - 변수처럼 생각해서 내가 원하는 요약 통계치 값을 변수에 넣고 추출
- case 2
  - 전체 데이터 내에서 판다스 문법으로 출력하는 경우
- 2개 이상도 위와 동일하게 추출 가능
---
- 2개 이상의 요약 통계치를 보고 싶은 경우
- **agg 함수 같이 사용**
- 데이터프레임.groupby('내가 원하는 컬럼들').agg((변수명1=('내가 요약 통계치 보고싶은 컬럼1', '요약 통계치1',
(변수명2=('내가 요약 통계치 보고싶은 컬럼2', '요약 통계치2') ... n)
- 내가 원하는 함수 만들어 적용 가능

## **cut, qcut**

- 구간화 binning 작업
  - 연속형 데이터를 고정된 간격 (bins) 으로 나눈다.
  - 수치형 데이터를 범주형 데이터로 구간화 변환
- **cut**
  - 구간 크기 일정하게, 값의 분포를 고려하지 않는다. 구간 개수는 직접 설정 가능하다.
  - pd.cut(데이터프레임 컬럼, bins=지정)
- **qcut**
  - 가능한 데이터의 분포에 따라 자동으로 분위 수 (Quantile) 기준을 나누는 게 큰 차이
  - 데이터의 개수가 비슷한 그룹으로 나누고 싶을 때
  - 등 간격이 아니라 분포를 고려해서 구간을 자동 설정
  - 구간에 속하는 데이터 개수가 동일하다.
  - pd.qcut(데이터프레임 컬럼, q=지정)

```python
tt['cut']=pd.cut(tt['age'], bins=4)
```

```python
tt['qcut']=pd.qcut(tt['age'], q=4)
```

```python
tt['age']
```

```python
tt.cut.value_counts()
```

```python
tt.qcut.value_counts()
```

## **Q 타이타닉 데이터를 확인했을 때**

- 등급이 존재하고, 생존에 대한 생존자 확인 가능
- 타이타닉 데이터의 생존자와 비생존자의 등급 별 생존자 차이가 유의미한가? 유의미하지 않은가?

```python
tt.groupby('pclass')['survived'].sum()
```

- pclass 클래스와 생존에 따른 분석
- 범주형 형태로 카이제곱 통계량으로 분석 가능

```python
crss_table=pd.crosstab(tt['pclass'], tt['survived'])
```

```python
crss_table
```

```python
# scipy 에서 제공하는 chi2_contingency 함수로 쉽게 추출 가능

from scipy import stats

chi2, p, dof, expected = stats.chi2_contingency(crss_table)
```

```python
# p-value

if p < 0.05:
  print('pclass, 생존율 차이는 통계적으로 유의미하다.')
else:
  print('pclass, 생존율 차이는 통계적으로 유의미하지 않다.')
```

## **카이제곱 간단한 정리**

- 두 개 이상의 범주형 변수 관계를 통계적 방법으로 분석
- 카이제곱 검정의 주요 용도
  - 독립성: 두 범주형 변수 간의 연관성 여부를 확인하는 것
    - 등급 (1, 2, 3) 에 따라 생존여부 (0, 1) 통계적으로 유의미한 차이가 있는지?
  - 적합성: 관찰된 데이터가 기대되는 분포와 일치하는지 여부 평가
    - 주사위 던지기
    - 1/6 기대빈도: 실제 주사위를 던지면서 나오는 관측된 빈도와 일치하는지 분석
  - 카이제곱 검정 가설
    - 귀무가설 (H0): 두 변수는 독립적이다. (연관성 X)
    - 대립가설 (H1): 두 변수는 독립적이지 않다. (연관성 O)
  - 위에 추출된 값들을 확인하면
    - 관찰빈도
    - 기대빈도
    - 통계량
    - p-value
    - 결론: p-value (0.05) 보다 작으면 두 변수 간 연관성 O
