## **데이터 분석 기초문법**

- loc[], iloc[] 로케이션, 인덱스 위치
  - loc - location (인덱스), 처음 생성된 데이터프레임의 인덱스
  - iloc - Integer location (넘버링된 인덱스), 현재 기준의 데이터프레임 순서, 데이터프레임을 재구조화 하면 그 기준으로 넘버링
  - loc[인덱스]
  - iloc[인덱스]
  - 안되는 조건
    - loc[행, 열], iloc[행, 열]
    - loc[:,:], iloc[:,:]
    - loc, 명시적으로 컬럼을 적어야 한다
    - iloc, 숫자로만 컬럼을 적어야 한다
- query()
- df[df[불린]]
- sort_values()
- 파생변수 - 판다스 문법, assign, lambda
- groupby
- merge, concat
- str
- 기타 stats 패키지 함수

```python
tt.loc[5:15, 'survived']
```

```python
tt.loc[5:15, ['survived', 'pclass']]
```

```python
# tt.iloc[5:15, 'survived']
tt.iloc[5:15, 0:1]
```

```python
# 인덱스 번호 0~100 A집단, 101~200 B집단
# A집단의 평균 나이와 B집단 평균 나이 분석해보자
seg_a_avg=tt.loc[0:100, 'age'].mean()
seg_b_avg=tt.loc[101:200, 'age'].mean()
```

```python
seg_b_avg
```

- query 문법
  - query(조건)
  - query('변수 연산자 조건')
    - 주의할 점은 문자열은 "'조건연산자 '문자열'"

```python
# pclass 3번과 pclass 1번 고객들의 남성, 여성 각각 fare 평균을 나눠서 구하여라
pc3_male=tt[['pclass', 'fare', 'sex']].query("pclass==3 & sex=='male'")['fare'].mean()
```

```python
pc3_male
```

```python
tt[['pclass', 'fare', 'sex']].query("pclass==1 & sex=='male'")['fare'].mean()
```

```python
tt[['pclass', 'fare', 'sex']].query("pclass==3 & sex=='female'")['fare'].mean()
```

```python
tt[['pclass', 'fare', 'sex']].query("pclass==1 & sex=='female'")['fare'].mean()
```

- df[df 비교연산자 불린]
  - 안에 들어가는 불린 값이 참인 조건의 데이터만 추출
- 다중조건
  - df[(df 조건 1) & (df 조건 2)]

```python
tt
```

```python
# 20대 이상, pclass==3 인 경우를 구해보자
# tt[(tt['age']>20)] & (tt['pclass]==3)][['fare', 'sex', 'pclass']].loc[0:100]['fare'].mean()
tt[(tt['age']>20) & (tt['pclass']==3)][['fare', 'sex', 'pclass']].loc[0:100]['fare']
```

```python
# loc, iloc 값 차이 남
# tt[(tt['age']>20)] & (tt['pclass]==3)][['fare', 'sex', 'pclass']].iloc[0:100]['fare'].mean()
tt[(tt['age']>20) & (tt['pclass']==3)][['fare', 'sex', 'pclass']].iloc[0:100]['fare']
```

```python
tt[(tt['age']>20) & (tt['pclass']==3)][['fare', 'sex', 'pclass']]
```

```python
# 데이터프레임 기준 필요
tt[['fare', 'age']].sort_values(by='fare', ascending=False)
```
